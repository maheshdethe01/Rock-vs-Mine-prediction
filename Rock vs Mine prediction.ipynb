#### Importing the necessary librbaries

```python
import numpy as np # for array
import pandas as pd # for loading our data 
from sklearn.model_selection import train_test_split # for splitting the data
from sklearn.linear_model import LogisticRegression # model
from sklearn.metrics import accuracy_score # for checking the accuracy of the model
```

#### Data Collection and Processing

```python
# loading the data
file_path = r'C:\Users\mahes\OneDrive\Documents\MAHESH\PROJECTS\ROCK MINE Prediction\Sonar data.csv'
data = pd.read_csv(file_path, header = None)
```

```python
# let's have a look into our dataset
data.head()
```

```python
# size of the data
data.shape
```

```python
data.describe()
```

```python
# see if the data is balanced ( same number of R and M (0's and 1's) )
data[60].value_counts()
# we can see that we have almost equal no. of predictive values categories
# R means Rock and M means Mine
```

```python
# seperating features and dependent variable
X = data.drop(columns = 60, axis = 1)
y = data[60]
```

#### Train-Test Data

```python
# splitting the data into train data and test data
X_train, X_test, y_train , y_test = train_test_split(X,y, train_size= 0.9, test_size= 0.1, stratify= y, random_state= 1)
```

```python
# checking the shape of the data 
print(X_train.shape[0],X_test.shape[0],y_train.shape[0],y_test.shape[0])
```

#### Model Training

```python
model = LogisticRegression()
```

```python
# fit the training data into the model 
model.fit(X_train, y_train)
```

#### Model Evaluation 

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
# accuracy score 
print(train_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
# accuracy score
print(test_data_accuracy)
```
